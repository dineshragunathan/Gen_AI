To plot the Heart shaped graph using mathematical functions , I figured out that the upper halves would be a tweaked version of a circle equation , but I had no idea about the lower halves , so I refered the internet are went through a few resources and found one particular resource to be very useful ( ref : https://medium.com/towards-data-science/plot-the-shape-of-my-heart-698d4776c51a ). I have used the two main equations to plot this figure , #1 equation of a circle and #2 equation of cos inverse as given in the article for this purpose. The cos inverse function fit exactly like it was required.

```python
import matplotlib.pyplot as plt
import numpy as np

''' Upper Part'''
" Equation of circle (tweaked)"

x = np.linspace(-3,3,1000)
y = (1 - ((abs(x) - 1)**2))**1/2

plt.figure(1)
plt.plot(x ,y)
plt.axvline(x=0,linestyle="--",color="gray")
plt.show()
```

```python
import math

''' Bottom Part'''
"Equation of cos inverse"

x = np.linspace(-2,2,1000)

''' I could not directly use the cos inverse equation in the plot , especially the y cordinate , as the cos inverse function returns a singular value , so I had to make a function that returns the value and had to vectorize the function.'''

def f(x):
    return math.acos(1 - abs(x)) - math.pi

fy = np.vectorize(f)
plt.figure(1)
plt.plot(x ,fy(x))
plt.axvline(x=0,linestyle="--",color="gray")
plt.show()
```

```python
'''Combining both the halves'''

x = np.linspace(-2,2,10000)
y = (1 - ((abs(x) - 1)**2))**1/2

plt.figure(1)
plt.plot(x ,y)
plt.axvline(x=0,linestyle="--",color="gray")

x2 = np.linspace(-2,2,10000)

def f(x):
    return math.acos(1 - abs(x)) - math.pi

fy = np.vectorize(f)
plt.plot(x2 ,fy(x2))
plt.axvline(x=0,linestyle="--",color="gray")

plt.show()

```
